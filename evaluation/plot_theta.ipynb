```python
import numpy as np
import sys
import math
from scipy.stats import norm

%matplotlib inline
import matplotlib.colors
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.mlab import griddata

sys.path.append('..')
from higgs_inference import settings
```

```python
benchmarks = [settings.theta_benchmark_nottrained, settings.theta_benchmark_trained]
denoms = [settings.theta1_default] + settings.theta1_alternatives
observed = [settings.theta_observed]
all_vip = observed + benchmarks + denoms

plt.figure(figsize=(12,12))

for i, theta in enumerate(settings.thetas):
    plt.text(theta[0],theta[1],str(i),
             fontsize = 16 if i in all_vip else 10,
             alpha = 1. if i in all_vip else 0.8 if i in settings.extended_pbp_training_thetas else 0.2,
             color = 'green' if i in observed else 'blue' if i in benchmarks else 'red' if i in denoms else 'black',
             horizontalalignment = 'center',
             verticalalignment = 'center')
    
plt.xlim(-1.03,1.03)
plt.ylim(-1.03,1.03)
plt.xlabel(r"$f_{W} \, v^2 / \Lambda^2$")
plt.ylabel(r"$f_{WW} \, v^2 / \Lambda^2$")

plt.tight_layout()
plt.savefig('../figures/debug/thetas.pdf')
```
