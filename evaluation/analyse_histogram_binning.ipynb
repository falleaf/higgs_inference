```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import sys
import math
import numpy as np
import collections

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt

sys.path.append('../')
from higgs_inference import settings
from higgs_inference.various.utils import s_from_r

```

```python
X = np.load('../data/unweighted_events/X_illustration.npy')
y = np.load('../data/unweighted_events/y_illustration.npy')
t = np.load('../data/unweighted_events/scores_illustration.npy')
r = np.load('../data/unweighted_events/r_illustration.npy')

pt = X[:,1]
deltaphi = X[:,41]

r_hat_coarse = np.load('../results/histo/r_illustration_histo_2d_roughbinning.npy')
r_hat_standard = np.load('../results/histo/r_illustration_histo_2d.npy')
r_hat_fine = np.load('../results/histo/r_illustration_histo_2d_finebinning.npy')
r_hat_superfine = np.load('../results/histo/r_illustration_histo_2d_superfinebinning.npy')
```

```python
plt.figure(figsize=(10.,8.))

skip = 1
rmin, rmax = -2.,2.



ax = plt.subplot(2,2,1)

plt.scatter(pt[::skip], deltaphi[::skip], c=np.log(r_hat_coarse)[::skip], cmap='viridis', vmin=rmin, vmax=rmax, s=5.)
cbar = plt.colorbar()

plt.xlabel(r'$p_{T,j1}$ [GeV]')
plt.ylabel(r'$\Delta \phi_{jj}$')
cbar.set_label(r'$\hat{r}(x | \theta_{SM})$ (coarse)')
plt.xlim(20.,1000.)
plt.ylim(0., math.pi)



ax = plt.subplot(2,2,2)

plt.scatter(pt[::skip], deltaphi[::skip], c=np.log(r_hat_standard)[::skip], cmap='viridis', vmin=rmin, vmax=rmax, s=5.)
cbar = plt.colorbar()

plt.xlabel(r'$p_{T,j1}$ [GeV]')
plt.ylabel(r'$\Delta \phi_{jj}$')
cbar.set_label(r'$\hat{r}(x | \theta_{SM})$ (standard)')
plt.xlim(20.,1000.)
plt.ylim(0., math.pi)



ax = plt.subplot(2,2,3)

plt.scatter(pt[::skip], deltaphi[::skip], c=np.log(r_hat_fine)[::skip], cmap='viridis', vmin=rmin, vmax=rmax, s=5.)
cbar = plt.colorbar()

plt.xlabel(r'$p_{T,j1}$ [GeV]')
plt.ylabel(r'$\Delta \phi_{jj}$')
cbar.set_label(r'$\hat{r}(x | \theta_{SM})$ (fine)')
plt.xlim(20.,1000.)
plt.ylim(0., math.pi)



ax = plt.subplot(2,2,4)

plt.scatter(pt[::skip], deltaphi[::skip], c=np.log(r_hat_superfine)[::skip], cmap='viridis', vmin=rmin, vmax=rmax, s=5.)
cbar = plt.colorbar()

plt.xlabel(r'$p_{T,j1}$ [GeV]')
plt.ylabel(r'$\Delta \phi_{jj}$')
cbar.set_label(r'$\hat{r}(x | \theta_{SM})$ (very fine)')
plt.xlim(20.,1000.)
plt.ylim(0., math.pi)


plt.tight_layout()
plt.show()

```
