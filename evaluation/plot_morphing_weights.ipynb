# Morphing weights

```python
# Morphing
n_samples = 15

def calculate_wtilde(t, component_sample):
    wtilde_components = np.asarray([
        1. + 0. * t[:, 0],
        t[:, 1],
        t[:, 1] * t[:, 1],
        t[:, 1] * t[:, 1] * t[:, 1],
        t[:, 1] * t[:, 1] * t[:, 1] * t[:, 1],
        t[:, 0],
        t[:, 0] * t[:, 1],
        t[:, 0] * t[:, 1] * t[:, 1],
        t[:, 0] * t[:, 1] * t[:, 1] * t[:, 1],
        t[:, 0] * t[:, 0],
        t[:, 0] * t[:, 0] * t[:, 1],
        t[:, 0] * t[:, 0] * t[:, 1] * t[:, 1],
        t[:, 0] * t[:, 0] * t[:, 0],
        t[:, 0] * t[:, 0] * t[:, 0] * t[:, 1],
        t[:, 0] * t[:, 0] * t[:, 0] * t[:, 0]
    ]).T
    return wtilde_components.dot(component_sample)


def calculate_wi(t, component_sample, sigma_sample, sigma_component):
    wtildes = calculate_wtilde(t, component_sample)
    sigma_wtildes = sigma_sample * wtildes # (?, 15)
    
    denom = np.ones_like(sigma_wtildes.T) # (?, 15)
    denom[:,:] /= np.sum(sigma_wtildes, axis=1)
    denom = denom.T
    
    return (sigma_wtildes * denom).T
```

# Errors from morphing

```python
delta_ri = 1.

uncertainty_on_r_from_morphing = np.zeros_like(wi_used[0])

for i, wi in enumerate(wi_used):
    uncertainty_on_r_from_morphing += wi**2 * delta_ri**2
    
uncertainty_on_r_from_morphing = np.sqrt(uncertainty_on_r_from_morphing)

print(np.min(uncertainty_on_r_from_morphing))
```

```python
delta_rij = 1.

uncertainty_on_r_from_decomposition = np.zeros_like(wi_used[0])

for i, wi in enumerate(wi_used):
    
    numerator = np.zeros_like(wi_used[0])
    for j, wj in enumerate(wi_theta1_used):
        numerator += wj**2 / wi**2 * delta_rij**2
        
    denominator = np.zeros_like(wi_used[0])
    for j, wj in enumerate(wi_theta1_used):
        denominator += wj / wi * 1
    denominator = denominator**4
    
    uncertainty_on_r_from_decomposition += numerator / denominator
```

```python
fig = plt.figure(figsize=(5.,4.))
ax = plt.gca()
    
pcm = ax.pcolormesh(xi, yi, uncertainty_on_r_from_morphing,
                   norm=matplotlib.colors.LogNorm(vmin=0.1, vmax=100.0),
                   cmap='Reds')
cbar = fig.colorbar(pcm, ax=ax, extend='both')

plt.scatter(thetas_used_morphing[:, 0], thetas_used_morphing[:, 1],
            marker='o', c='black', s=25, lw=0, zorder=9, alpha=0.3)

plt.xlim(-1.05, 1.05)
plt.ylim(-1.05, 1.05)
plt.xlabel(r"$f_{W} \, v^2 / \Lambda^2$")
plt.ylabel(r"$f_{WW} \, v^2 / \Lambda^2$")
cbar.set_label(r'$\Delta r \, / \, \Delta r_i$')

plt.tight_layout()
plt.savefig(figure_dir + "/morphing_uncertainties.pdf")
```

```python
fig = plt.figure(figsize=(5.,4.))
ax = plt.gca()
    
pcm = ax.pcolormesh(xi, yi, uncertainty_on_r_from_decomposition,
                   norm=matplotlib.colors.LogNorm(vmin=0.1, vmax=10000.0),
                   cmap='Reds')
cbar = fig.colorbar(pcm, ax=ax, extend='both')

plt.scatter(thetas_used_morphing[:, 0], thetas_used_morphing[:, 1],
            marker='o', c='black', s=25, lw=0, zorder=9, alpha=0.3)

plt.xlim(-1.05, 1.05)
plt.ylim(-1.05, 1.05)
plt.xlabel(r"$f_{W} \, v^2 / \Lambda^2$")
plt.ylabel(r"$f_{WW} \, v^2 / \Lambda^2$")
cbar.set_label(r'$\Delta r \, / \, \Delta r_{ij}$')

plt.tight_layout()
plt.savefig(figure_dir + "/double_decomposition_uncertainties.pdf")
```
